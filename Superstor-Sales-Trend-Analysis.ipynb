***This converts the 'Order Date' column into a real date that pandas can understand.***

```python
df['Order Date']=pd.to_datetime(df["Order Date"])
df['Ship Date']=pd.to_datetime(df["Ship Date"])
```

```python
df
```

**.dt.month_name gives you month**

**.dt.year gives you year**      

**.dt.to_period("m") gives you year nd month**

```python
df["Order Month"] = df["Order Date"].dt.month_name()   # e.g., January
df["Order Year"] = df["Order Date"].dt.year            # e.g., 2023
df["Month-Year"] = df["Order Date"].dt.to_period("M")  # e.g., 2023-01
```

***How do sales and profit vary each month?***

```python
df[['Order Date','Month-Year']].head()
```

**->This gives you a new DataFrame with:**

**Each month**

**Total sales for that month**

**Total profit for that month**

```python
monthly_data = df.groupby('Month-Year')[['Sales', 'Profit']].sum().reset_index()
df
```

```python
monthly_data = monthly_data.sort_values('Month-Year')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))    #This is used to show how big the graph need to be
plt.plot(monthly_data['Month-Year'].astype(str), monthly_data['Sales'], label='Sales', marker='o')   #marker is used to tick, convert thm into str for good display,month-year on x-axis and sales and profits on y axis
plt.plot(monthly_data['Month-Year'].astype(str), monthly_data['Profit'], label='Profit', marker='s')
plt.xticks(rotation=45)   #it rotates the x-axis coz they are bigg if we didnt rotate they are not visible they will collide
plt.xlabel('Month-Year')
plt.ylabel('Amount')
plt.title('Monthly Sales and Profit Trend')
plt.legend()    #Displays a legend to show which line is Sales and which is Profit.
plt.tight_layout()    #Adjusts spacing automatically to prevent labels from getting cut off
plt.show()
```

****Which month generated the most sales?****

```python
monthly_data.loc[monthly_data['Sales'].idxmax()]
```

****Which month had the lowest profit?****

```python
monthly_data.loc[monthly_data['Sales'].idxmin()]
```

```python
monthly_data[(monthly_data['Sales'] > monthly_data['Sales'].mean()) & 
             (monthly_data['Profit'] < monthly_data['Profit'].mean())]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x positions for each bar
x = np.arange(len(monthly_data['Month-Year']))
width = 0.35  # width of the bars

# Set figure size
plt.figure(figsize=(14, 6))

# Plot Sales and Profit side by side
plt.bar(x - width/2, monthly_data['Sales'], width, label='Sales', color='skyblue')
plt.bar(x + width/2, monthly_data['Profit'], width, label='Profit', color='salmon')

# Label x-axis with Month-Year
plt.xticks(x, monthly_data['Month-Year'], rotation=45)

# Add labels and title
plt.xlabel('Month-Year')
plt.ylabel('Amount in ₹')
plt.title('Monthly Sales vs Profit')
plt.legend()

# Clean layout
plt.tight_layout()    #Fixes overlaps between labels and figure edges.

# Show the chart
plt.show()
```
